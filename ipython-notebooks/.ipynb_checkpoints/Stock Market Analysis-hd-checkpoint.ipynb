```python
# stock risk analysis
# make new clean version of tech_ret dataframe
# compare standard deviation of daily return with expected return
rets = air_rets.dropna()
```


The 0.05 empirical quantile of daily returns is at "-0.0301". That means that with "95"% confidence, our worst daily loss will not exceed 3%. If we have a 1 million dollar investment, our one-day 5% VaR is -0.0301 * 1,000,000 = $30,100.

```python
# Now we'll use the Monte Carlo method
# We use GBM (Geometric Brownian Motion)
# Check this link http://www.investopedia.com/articles/07/montecarlo.asp
```

```python
# now let's do a few simulations
# set up the time horizon
days = 365

# delta
dt = 1/days

# mu is drift from the expected return data above
mu = rets.mean()['UAL']

# volatility from the STD of the average return
sigma = rets.std()['UAL']
```

```python
# create a function
def stock_monte_carlo(start_price,days,mu,sigma):
    # define price array
    price = np.zeros(days)
    price[0] = start_price
    
    # define shock and drift arrays
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    # run price array for the number of days
    for x in xrange(1,days):
        
        # calc shock
        shock[x] = np.random.normal(loc=mu*dt,scale= sigma*np.sqrt(dt))
        
        # calc drift
        drift[x] = mu * dt
        
        # calc price
        price[x] = price[x-1]+ (price[x-1] * (drift[x] + shock[x]))
    
    return price
```

```python
UAL.head()
```

```python
start_price = 563.56

# now run the simulation 100 times
for run in xrange(100):
    plt.plot(stock_monte_carlo(start_price,days,mu,sigma))
    
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Monte Carlo Analysis for United Airlines')
```

```python
# now create a histogram with these results
# do 10000 runs
runs = 10000

simulations = np.zeros(runs)

for run in xrange(runs):
    simulations[run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1]
```

```python
# def q as the 1% empirical quantile
q = np.percentile(simulations,1)

# plot distribution of the end prices
plt.hist(simulations,bins=200)

# starting price
plt.figtext(0.6,0.8, s="Start price: $%.2f" %start_price)
# mean ending price
plt.figtext(0.6,0.7, "Mean final price: $%.2f" % simulations.mean())

# Variance of the price (within 99% confidence interval)
plt.figtext(0.6, 0.6, "VaR(0.99): $%.2f" % (start_price - q,))

# Display 1% quantile
plt.figtext(0.15, 0.6, "q(0.99): $%.2f" % q)

#plot a line at the 1% quantile result
plt.axvline(x=q, linewidth=4, color='r')

# Title
plt.title(u"Final price distribution for United Air Stock after %s days" % days, weight='bold');
```

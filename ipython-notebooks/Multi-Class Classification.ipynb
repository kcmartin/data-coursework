```python
# data imports
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

# plot imports
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# load the dataset from SciKit learn
from sklearn import linear_model
from sklearn.datasets import load_iris

# import the data
iris = load_iris()

#Grab features (X) and the Target (Y)
X = iris.data

Y = iris.target

#show the builtin data description
print iris.DESCR
```

```python
# put the data into a DataFrame
iris_data = DataFrame(X,columns=['Sepal Length','Sepal Width','Petal Length','Petal Width'])
```

```python
iris_target = DataFrame(Y,columns=['Species'])
```

```python
iris_target.head()
```

```python
# need to rename the species to names instead of numeric categories
# review 'split apply combine' lecture
# permanently change the Species column by applying the function

# function to rename
def flower(num):
    if num == 0:
        return 'Setosa'
    elif num == 1:
        return 'Versicolor'
    else:
        return 'Viginica'
```

```python
iris_target['Species'] = iris_target['Species'].apply(flower)
```

```python
iris_target.head()
```

```python
# concatenate the two dataframes
iris = pd.concat([iris_data,iris_target],axis=1)
```

```python
iris.head()
```

```python
# now some visualization
sns.pairplot(iris,hue='Species',size=2)
```

```python
# we can see that Setosa is distinct from the other two species
```

```python
sns.factorplot('Petal Length',data=iris,hue='Species',size=8)
```
